```python
import requests
import pandas as pd
from io import StringIO
import datetime
import time
import os
def crawl_monthly_report(date):
    
    url=f'https://mops.twse.com.tw/nas/t21/sii/t21sc03_{str(date.year-1911)}_{str(date.month)}_0.html'
    
    if date.year-1911 <= 98:
        url=f'https://mops.twse.com.tw/nas/t21/sii/t21sc03_{str(date.year-1911)}_{str(date.month)}.html'
        
    
    # 偽瀏覽器
#     headers={'User-Agent':'Mozilla/5.0'}

    r = requests.get(url)
    r.encoding = 'big5'
    
    try:
        html_df = pd.read_html(StringIO(r.text))
    except:
        return None
    

    # 將dfs中，row長度介於5~11的table合併（這些才是我們需要的table，其他table不需要）
    df = pd.concat([df for df in html_df if df.shape[1] <= 11 and df.shape[1] > 5])

    # 設定column名稱
    df.columns = df.columns.get_level_values(1)

    # 將 df 中的當月營收用 .to_numeric 變成數字，再把其中不能變成數字的部分以 NaN 取代（errors='coerce'）
    df['當月營收'] = pd.to_numeric(df['當月營收'], 'coerce')

    # 再把當月營收中，出現 NaN 的 row 用 .dropna 整行刪除
    df = df[~df['當月營收'].isnull()]

    # 刪除「公司代號」中出現「合計」的行數，其中「～」是否定的意思
    df = df[df['公司代號'] != '合計']

    # 將「公司代號」與「公司名稱」共同列為 df 的 indexes
    df = df.set_index(['公司代號', '公司名稱'])
    
    if not os.path.isdir(f'./monthly_report/{date.year}'):
        os.mkdir(f'./monthly_report/{date.year}')
        
    # 把 df 存成 csv 檔，並且命名為「test.csv」，指定用「utf_8_sig」編碼
    df.to_csv(f'./monthly_report/{date.year}/{date.month}.csv', encoding='utf_8_sig')
    print(f'{date.year}{date.month}')
    return df

# def mkadir(year):
#     os.mkdir(f'./mon')

# df.head()
```

```python
for j in range(10):
    for i in range(12):
        crawl_monthly_report(datetime.date(j+2012,i+1,26))
        time.sleep(3)
```
